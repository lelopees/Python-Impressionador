### Outros tipos de Gráfico:

- Linha

```python
# min, max, e qtde
vendas = np.random.randint(1000, 3000, 50)
# inicio, fim 
meses = np.arange(1, 51)
```

```python
#print(len(vendas))
#print(len(meses))

plt.plot(meses,vendas, color = 'grey')
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Editando o Gráfico de Linha

```python
plt.plot(meses,vendas)
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
plt.scatter(meses,vendas) 
plt.show()
```

```python
#grafico de barras
plt.bar(meses,vendas) 
plt.show()
```

```python
plt.figure(figsize=(15,3))

# qtde de linhas, qtde de graficos, qual a posição desse grafico
# pode ser 131 tbm
plt.subplot(1,3,1)
plt.bar(meses,vendas) 

plt.subplot(1,3,2)
plt.plot(meses,vendas, color = 'red')

plt.subplot(1,3,3)
plt.scatter(meses,vendas) 

plt.show()
```
